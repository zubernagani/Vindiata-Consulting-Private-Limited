**Step 2 :**

**Filtering the data by Slot S1 and S2**


```python
# Function to filter data by slot
def get_slot_data(df, date, slot):
    """Filters data for a given date and slot."""
    # Convert 'date' to a pd.Timestamp object
    date = pd.Timestamp(date)
    
    if slot == 'S1':  # Slot S1: 12 am - 11:59 am
        start, end = date.replace(hour=0, minute=0), date.replace(hour=11, minute=59)
    elif slot == 'S2':  # Slot S2: 12 pm - 11:59 pm
        start, end = date.replace(hour=12, minute=0), date.replace(hour=23, minute=59)
    
    # Filter data within the given time range
    return df[(df['Datetime'] >= start) & (df['Datetime'] <= end)]
```

**Step 3:**

**Create a function that calculates Loyalty Points with the formula given**


```python
# Function to calculate loyalty points
def calculate_loyalty_points(user_id, games_played, deposit_amount, withdrawal_amount, deposit_count, withdrawal_count):
    points = (0.01 * deposit_amount) + \
             (0.005 * withdrawal_amount) + \
             (0.001 * max(deposit_count - withdrawal_count, 0)) + \
             (0.2 * games_played)
    return points
```

**Question No: 1**

**Find Playerwise Loyalty points earned by Players in the following slots:-**

**2nd October Slot S1**

**16th October Slot S2**

**18th October Slot S1**

**26th October Slot S2**
    


```python
slot_info = [
    {'date': '2022-10-02', 'slot': 'S1'},
    {'date': '2022-10-16', 'slot': 'S2'},
    {'date': '2022-10-18', 'slot': 'S1'},
    {'date': '2022-10-26', 'slot': 'S2'}
]

results = []
```

**Question No:2**


**Calculate overall loyalty points earned and rank players on the basis of loyalty points in the month of October.** 




```python
# Loop through each date and slot combination to calculate loyalty points
for slot_detail in slot_info:
    # Extract date and slot information
    date_str = slot_detail['date']
    slot = slot_detail['slot']
    date = pd.to_datetime(date_str, format='%Y-%m-%d')
    
    # Filter gameplay, deposit, and withdrawal data for the current slot
    games_data = get_slot_data(gameplay_data, date, slot)
    deposit_data_slot = get_slot_data(deposit_data, date, slot)
    withdrawal_data_slot = get_slot_data(withdrawal_data, date, slot)
    
    # Identify unique user IDs across all data types for the slot
    unique_user_ids = set(games_data['User ID']).union(deposit_data_slot['User Id'], withdrawal_data_slot['User Id'])
    
    # Calculate loyalty points for each user in the current slot
    for user_id in unique_user_ids:
        # Sum up games played, deposit amounts, and withdrawal amounts
        total_games_played = games_data[games_data['User ID'] == user_id]['Games Played'].sum()
        total_deposit_amount = deposit_data_slot[deposit_data_slot['User Id'] == user_id]['Amount'].sum()
        total_withdrawal_amount = withdrawal_data_slot[withdrawal_data_slot['User Id'] == user_id]['Amount'].sum()
        
        # Count deposit and withdrawal occurrences
        deposit_transactions = deposit_data_slot[deposit_data_slot['User Id'] == user_id].shape[0]
        withdrawal_transactions = withdrawal_data_slot[withdrawal_data_slot['User Id'] == user_id].shape[0]
        
        # Compute loyalty points using the provided formula
        loyalty_points = calculate_loyalty_points(
            user_id, total_games_played, total_deposit_amount, 
            total_withdrawal_amount, deposit_transactions, withdrawal_transactions
        )
        
        # Append the results to the list
        results.append({
            'User Id': user_id, 
            'Date': date_str, 
            'Slot': slot, 
            'Loyalty Points': loyalty_points
        })
```

```python
# Convert results to DataFrame
results_df = pd.DataFrame(results)
print("Loyalty Points for specified slots:\n", results_df)
```

```python

# Define the date range for October
october_start = pd.Timestamp('2022-10-01')
october_end = pd.Timestamp('2022-10-31 23:59')
```

```python

games_october = gameplay_data.query('@october_start <= Datetime <= @october_end')
deposits_october = deposit_data.query('@october_start <= Datetime <= @october_end')
withdrawals_october = withdrawal_data.query('@october_start <= Datetime <= @october_end')
```

```python
# Initialize a dictionary to store loyalty points and games played for each user
user_points = {}
```

```python

# Set of all unique user IDs in October data
unique_users = set(games_october['User ID']).union(deposits_october['User Id'], withdrawals_october['User Id'])

# Loop over each unique user and calculate loyalty points
for user_id in unique_users:
    # Aggregate games played, deposit amount, and withdrawal amount for each user
    games_played = games_october.query('`User ID` == @user_id')['Games Played'].sum()
    deposit_amount = deposits_october.query('`User Id` == @user_id')['Amount'].sum()
    withdrawal_amount = withdrawals_october.query('`User Id` == @user_id')['Amount'].sum()
    
    # Count the number of deposit and withdrawal transactions
    deposit_count = deposits_october.query('`User Id` == @user_id').shape[0]
    withdrawal_count = withdrawals_october.query('`User Id` == @user_id').shape[0]
    
    # Compute loyalty points
    points = calculate_loyalty_points(
        user_id, games_played, deposit_amount, withdrawal_amount, deposit_count, withdrawal_count
    )
    
    # Store the results in the dictionary
    user_points[user_id] = {
        'Loyalty Points': points,
        'Games Played': games_played
    }
```

```python
# Convert the user_points dictionary into a DataFrame
user_points_df = pd.DataFrame.from_dict(user_points, orient='index').reset_index()

# Rename the 'index' column to 'User_Id'
user_points_df.rename(columns={'index': 'User_Id'}, inplace=True)

# Rank users by 'Loyalty Points' and use 'Games Played' as a tiebreaker
user_points_df.sort_values(by=['Loyalty Points', 'Games Played'], ascending=[False, False], inplace=True)

# Reset the index for a clean DataFrame
user_points_df.reset_index(drop=True, inplace=True)

# Display the ranked users
print("Ranked Users by Loyalty Points for October:\n", user_points_df)
```

**PART B - How much bonus should be allocated to leaderboard players**

```python
# Calculate loyalty points for each player
loyalty_points = {
    user_id: (
        (0.01 * deposit_data[deposit_data['User Id'] == user_id]['Amount'].sum()) +
        (0.005 * withdrawal_data[withdrawal_data['User Id'] == user_id]['Amount'].sum()) +
        (0.001 * max(deposit_data[deposit_data['User Id'] == user_id].shape[0] - withdrawal_data[withdrawal_data['User Id'] == user_id].shape[0], 0)) +
        (0.2 * gameplay_data[gameplay_data['User ID'] == user_id]['Games Played'].sum())
    )
    for user_id in set(gameplay_data['User ID']).union(deposit_data['User Id'], withdrawal_data['User Id'])
}

# Create DataFrame for loyalty points
loyalty_df = pd.DataFrame(list(loyalty_points.items()), columns=['User Id', 'Loyalty Points'])

# Identify top 50 players by loyalty points
top_50 = loyalty_df.nlargest(50, 'Loyalty Points')

# Calculate total loyalty points and total games played for top 50 players
total_loyalty_points = top_50['Loyalty Points'].sum()
total_games_played = gameplay_data[gameplay_data['User ID'].isin(top_50['User Id'])]['Games Played'].sum()

# Bonus pool allocation
total_bonus_pool = 50000
loyalty_weight = 0.7
games_weight = 0.3

top_50['Points Bonus'] = (top_50['Loyalty Points'] / total_loyalty_points) * total_bonus_pool * loyalty_weight
top_50['Games Bonus'] = (top_50['Loyalty Points'] / total_games_played) * total_bonus_pool * games_weight

# Calculate total bonus for each player
top_50['Total Bonus'] = top_50['Points Bonus'] + top_50['Games Bonus']

# Display the top players with their bonus allocations
print(top_50[['User Id', 'Loyalty Points', 'Total Bonus']])
```
